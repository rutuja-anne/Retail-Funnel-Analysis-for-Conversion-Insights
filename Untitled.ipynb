```python
import pandas as pd
import numpy as np 
```

```python
# Step 1: Read both sheets separately
df_1 = pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2009-2010")
df_2 = pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2010-2011")

# Step 2: Combine (append) both into one DataFrame
df = pd.concat([df_1, df_2], ignore_index=True)
```

```python
# Step 2: Clean the dataset properly

# Drop missing Customer IDs — we can’t do funnel/user analysis without this
df = df.dropna(subset=['Customer ID'])

# Drop missing descriptions (optional — just few)
df = df.dropna(subset=['Description'])

# Remove negative or zero Quantity and Price (returns or invalid)
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]

# Convert data types
df['Customer ID'] = df['Customer ID'].astype(int)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['Price']

print("✅ Cleaned Data Shape:", df.shape)
print(df.isnull().sum())
```

```python
# Stage 1: Browsed (at least one purchase)
stage1 = df['Customer ID'].nunique()

# Stage 2: High Intent (bought >=5 items at least once)
high_intent_customers = df[df['Quantity'] >= 5]['Customer ID'].unique()
stage2 = len(high_intent_customers)

# Stage 3: Repeat Customers (purchased > 1 time)
repeat_customers = df.groupby('Customer ID')['Invoice'].nunique()
stage3 = (repeat_customers > 1).sum()

# Create funnel dataframe
funnel_data = pd.DataFrame({
    'Stage': ['Browsed', 'High Intent (Bulk Buyers)', 'Repeat Customers'],
    'Users': [stage1, stage2, stage3]
})

# Drop-off calculation
funnel_data['Drop_off_%'] = funnel_data['Users'].pct_change().fillna(0) * -100
funnel_data


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(funnel_data['Stage'], funnel_data['Users'], marker='o', linewidth=2)
plt.title("Customer Funnel Analysis", fontsize=14)
plt.xlabel("Funnel Stage")
plt.ylabel("Number of Users")
plt.grid(True)
for i, val in enumerate(funnel_data['Users']):
    plt.text(i, val + 50, str(val), ha='center', fontsize=10)
plt.show()
```

```python
# Total Revenue
total_revenue = df['TotalPrice'].sum()

# Total Orders
total_orders = df['Invoice'].nunique()

# Total Customers
total_customers = df['Customer ID'].nunique()

# Average Order Value (AOV)
aov = total_revenue / total_orders

print(f"Total Revenue: £{total_revenue:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Total Customers: {total_customers}")
print(f"Average Order Value: £{aov:,.2f}")
```

```python
country_funnel = df.groupby('Country').agg({
    'Invoice': 'nunique',  # number of orders
    'Customer ID': 'nunique',
    'TotalPrice': 'sum'
}).reset_index()

country_funnel.rename(columns={
    'Invoice': 'Total_Orders',
    'Customer ID': 'Unique_Customers',
    'TotalPrice': 'Revenue'
}, inplace=True)

country_funnel['Avg_Order_Value'] = country_funnel['Revenue'] / country_funnel['Total_Orders']
country_funnel.sort_values('Revenue', ascending=False).head(10)
```

```python
# Remove rows where Description contains 'MANUAL'
df = df[~df['Description'].str.contains('MANUAL', case=False, na=False)]

print("✅ Removed invalid 'Manual' entries. Cleaned Shape:", df.shape)
```

```python
# Group by Customer and analyze their activity
funnel_df = df.groupby('Customer ID').agg({
    'Invoice': 'nunique',     # number of unique orders
    'Quantity': 'sum',        # total items bought
    'TotalPrice': 'sum'       # total money spent
}).reset_index()

# Create funnel stage flags
funnel_df['Browsed'] = 1
funnel_df['Added_to_Cart'] = (funnel_df['Quantity'] > 3).astype(int)
funnel_df['Purchased_Again'] = (funnel_df['Invoice'] > 1).astype(int)

# Count users at each stage
stages = {
    'Browsed': funnel_df['Browsed'].sum(),
    'Added to Cart': funnel_df['Added_to_Cart'].sum(),
    'Purchased Again': funnel_df['Purchased_Again'].sum()
}

# Calculate drop-off rates
funnel_data = pd.DataFrame({
    'Stage': list(stages.keys()),
    'Users': list(stages.values())
})
funnel_data['Drop_off_%'] = funnel_data['Users'].pct_change().fillna(0) * -100

print(funnel_data)
```

```python
import matplotlib.pyplot as plt

# Funnel chart
plt.figure(figsize=(8, 5))
plt.plot(funnel_data['Stage'], funnel_data['Users'], marker='o', linewidth=3)
plt.title('Customer Purchase Funnel', fontsize=14)
plt.xlabel('Funnel Stage')
plt.ylabel('Number of Users')
plt.grid(alpha=0.3)
for i, val in enumerate(funnel_data['Users']):
    plt.text(i, val + 50, str(val), ha='center', fontsize=10)
plt.show()
```
